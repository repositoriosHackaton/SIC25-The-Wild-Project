```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Crear una figura y un eje
fig, ax = plt.subplots(figsize=(10, 8))

# Definir las coordenadas y tamaños de los nodos
nodes = {
    "Inicio": (0.5, 0.9),
    "Generación de Contraseña": (0.5, 0.7),
    "Validación de Contraseña": (0.5, 0.5),
    "Fin": (0.5, 0.3)
}

# Dibujar los nodos
for node, (x, y) in nodes.items():
    ax.add_patch(patches.Rectangle((x-0.2, y-0.05), 0.4, 0.1, edgecolor='black', facecolor='lightblue'))
    plt.text(x, y, node, ha='center', va='center', fontsize=10)

# Dibujar las flechas
ax.arrow(0.5, 0.85, 0, -0.04, head_width=0.05, head_length=0.05, fc='k', ec='k')
ax.arrow(0.5, 0.65, 0, -0.04, head_width=0.05, head_length=0.05, fc='k', ec='k')
ax.arrow(0.5, 0.45, 0, -0.04, head_width=0.05, head_length=0.05, fc='k', ec='k')

# Configurar el aspecto del gráfico
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
from collections import Counter

# Datos de ejemplo: lista de contraseñas (podrías obtener esto de un archivo o base de datos)
passwords = [
    "123456", "password", "123456789", "12345678", "12345",
    "1234567", "1234567890", "qwerty", "abc123", "111111",
    "123456", "password", "123456789", "12345678", "12345",
    "1234567", "1234567890", "qwerty", "abc123", "111111",
    "123456", "password", "123456789", "12345678", "12345",
    "1234567", "1234567890", "qwerty", "abc123", "111111"
]

# Contar la frecuencia de cada contraseña
password_counts = Counter(passwords)

# Obtener las 10 contraseñas más comunes
most_common_passwords = password_counts.most_common(10)

# Separar las contraseñas y sus frecuencias
passwords_list, counts = zip(*most_common_passwords)

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(passwords_list, counts, color='skyblue')

# Añadir título y etiquetas
plt.title('Las 10 contraseñas más usadas', fontsize=16)
plt.xlabel('Contraseñas', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)

# Rotar las etiquetas del eje X para mejor visualización
plt.xticks(rotation=45, ha='right')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Definir los tipos de caracteres y su cantidad
tipos_caracteres = {
    "Solo números": 10,                     # 0-9 (10 caracteres)
    "Números y letras minúsculas": 36,       # 0-9 + a-z (36 caracteres)
    "Números, letras minúsculas y mayúsculas": 62,  # 0-9 + a-z + A-Z (62 caracteres)
    "Todos los caracteres imprimibles": 94   # 0-9 + a-z + A-Z + símbolos (94 caracteres)
}

# Longitud fija de la contraseña (por ejemplo, 8 caracteres)
longitud_contraseña = 8

# Calcular la entropía para cada tipo de caracteres
entropias = [np.log2(caracteres ** longitud_contraseña) for caracteres in tipos_caracteres.values()]

# Normalizar la entropía para obtener un porcentaje de seguridad (0% a 100%)
seguridad = (entropias / np.max(entropias)) * 100

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.bar(tipos_caracteres.keys(), seguridad, color='b', alpha=0.7, label='Seguridad de la contraseña')

# Personalizar la gráfica
plt.title(f'Porcentaje de Seguridad de Contraseñas (Longitud = {longitud_contraseña} caracteres)')
plt.xlabel('Tipo de Caracteres Utilizados')
plt.ylabel('Porcentaje de Seguridad (%)')
plt.ylim(0, 110)  # Ajustar el límite del eje Y
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')  # Rotar las etiquetas del eje X para mejor legibilidad
plt.legend()
plt.tight_layout()  # Ajustar el layout para que no se corten las etiquetas
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo dirigido
G = nx.DiGraph()

# Añadir nodos y conexiones para una red feed-forward simple
# Capa de entrada
input_layer = ['Input1', 'Input2', 'Input3']
# Capa oculta
hidden_layer = ['Hidden1', 'Hidden2']
# Capa de salida
output_layer = ['Output']

# Añadir nodos al grafo
G.add_nodes_from(input_layer, layer='input')
G.add_nodes_from(hidden_layer, layer='hidden')
G.add_nodes_from(output_layer, layer='output')

# Añadir conexiones
for input_node in input_layer:
    for hidden_node in hidden_layer:
        G.add_edge(input_node, hidden_node)

for hidden_node in hidden_layer:
    G.add_edge(hidden_node, output_layer[0])

# Dibujar el grafo
pos = nx.multipartite_layout(G, subset_key="layer")
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=10, font_weight='bold', arrows=True)
plt.title("Estructura de una Red Neuronal Prealimentada")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir funciones de activación
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Generar datos
x = np.linspace(-10, 10, 100)
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

# Graficar
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.plot(x, y_sigmoid, label='Sigmoid')
plt.title('Función Sigmoid')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(x, y_relu, label='ReLU', color='orange')
plt.title('Función ReLU')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(x, y_tanh, label='Tanh', color='green')
plt.title('Función Tanh')
plt.legend()

plt.tight_layout()
plt.show()
```
